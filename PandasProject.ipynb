## Data Cleaning

Standardize column names and values to improve consistency and prepare the dataset for analysis.

```python
df.rename({'AnimalName': 'Species'}, axis=1, inplace = True)
```

```python
df
```

```python
df['Species'] = df['Species'].replace('cat', 'Cat')
```

```python
df
```

### Create Derived Columns

Convert weight measurements from kilograms to pounds to make the data easier to interpret and analyze.

```python
df['Weight_lbs'] = (df['Weight_kg'] * 2.2046).round(1)
```

```python
df
```

### Reorganize Dataset Structure

Move the newly created Weight_lbs column to a more logical position within the dataset for readability.

```python
column_to_move = df.pop('Weight_lbs')
```

```python
df.insert(4, 'Weight_lbs', column_to_move)
```

```python
df
```

### Standardize Text Values

Apply consistent capitalization across categorical fields to eliminate duplicate categories caused by formatting differences.

```python
columns_to_capitalize = ['Species', 'Breed', 'MedicalHistory','Symptom_1','Symptom_2','Symptom_3','Symptom_4','Symptom_5']
df[columns_to_capitalize] = df[columns_to_capitalize].apply(lambda x: x.str.capitalize())
df[columns_to_capitalize]
```

### Improve Column Naming

Rename columns to follow a more consistent naming convention.

```python
df.rename({'MedicalHistory': 'Medical_History'}, axis=1, inplace = True)
```

```python
df
```

## Data Quality Assessment

Check the dataset for missing values and potential data quality issues.

```python
df.isnull().sum()
```

### Duplicate Record Review

Identify duplicate records that may impact reporting accuracy and investigate whether they represent true duplicates or valid repeated observations.

```python
df.duplicated().sum()
```

```python
duplicates = df[df.duplicated(keep=False)]
duplicates.head()
```

## Exploratory Data Analysis

Examine the distribution of animals by species to better understand the composition of the dataset.

```python
df['Species'].value_counts(normalize=True)
```

### Filter Data by Age

Review records for animals older than five years to explore age-related characteristics.

```python
df[df['Age'] > 5]
```

### Filter by Medical Conditions

Identify patients with selected medical histories for targeted analysis.

```python
df[df['Medical_History'].isin(['Allergies', 'Dental issues'])]
```

### Multi-Condition Filtering

Apply multiple filtering conditions to locate specific subsets of records.

```python
df[(df['Species'] == 'Dog') & (df['Symptom_1'] == 'Vomiting')]
```

### Average Age by Species

Calculate average age for each species using groupby aggregation.

```python
df.groupby('Species')['Age'].mean().round(0)
```

### Average Weight by Species

Compare average weight across species to identify differences between animal groups.

```python
df.groupby('Species')['Weight_lbs'].mean().round(2)
```

### Medical History Frequency

Determine the most common medical conditions represented in the dataset.

```python
df.groupby('Medical_History').size().sort_values(ascending=False)
```

### Most Common Breeds

Identify the most frequently occurring breeds within the dataset.

```python
df['Breed'].value_counts().head(10)
```
